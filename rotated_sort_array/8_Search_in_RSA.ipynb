There is an integer array nums sorted in ascending order (with distinct values).

Prior to being passed to your function, nums is possibly rotated at an unknown pivot index k (1 <= k < nums.length) such that the resulting array is [nums[k], nums[k+1], ..., nums[n-1], nums[0], nums[1], ..., nums[k-1]] (0-indexed). For example, [0,1,2,4,5,6,7] might be rotated at pivot index 3 and become [4,5,6,7,0,1,2].

Given the array nums after the possible rotation and an integer target, return the index of target if it is in nums, or -1 if it is not in nums.

You must write an algorithm with O(log n) runtime complexity.

 

Example 1:

Input: nums = [4,5,6,7,0,1,2], target = 0
Output: 4
Example 2:

Input: nums = [4,5,6,7,0,1,2], target = 3
Output: -1
Example 3:

Input: nums = [1], target = 0
Output: -1
 

Constraints:

1 <= nums.length <= 5000
-104 <= nums[i] <= 104
All values of nums are unique.
nums is an ascending array that is possibly rotated.
-104 <= target <= 104

Explanation:

Initialization: left and right pointers mark the boundaries of our search space.

Binary Search: We use binary search principles, repeatedly calculating the middle index (mid).

Target Found: If nums[mid] is the target, we return mid.

Checking Sorted Halves: We determine which half of the array is sorted by comparing nums[left] and nums[mid].

Adjusting Search Space: Based on where the target could potentially lie, we adjust left or right accordingly. This lets us narrow our focus to the relevant half of the array.

Target Not Found: If the loop finishes without finding the target, we return -1.

How the Code Works

The key to understanding this algorithm is the fact that in a rotated sorted array, one half will always be sorted. Once we determine the sorted half, we can apply typical 
binary search logic within that half.


Example: Consider nums = [4, 5, 6, 7, 0, 1, 2] and target = 0.


Initially, the left half is sorted.

Since target is in the range of the sorted half, we narrow our search to the left half.

We continue this, until we find the target at index 4.

Time Complexity: Each iteration of the loop cuts the search space in half, resulting in a logarithmic time complexity of O(log n).

```python
class Solution(object):
    def search(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """

        left, right = 0, len(nums) - 1

        while left <= right:
            mid = (left + right) // 2

            if nums[mid] == target:  # Target found
                return mid

            # Check if left half is sorted
            if nums[left] <= nums[mid]:  
                if nums[left] <= target < nums[mid]:
                    # Target lies in sorted left half
                    right = mid - 1
                else:
                    # Target lies in the right half
                    left = mid + 1

            # Otherwise, the right half is sorted
            else:  
                if nums[mid] < target <= nums[right]:
                    # Target lies in the sorted right half
                    left = mid + 1
                else:
                    # Target lies in the left half
                    right = mid - 1

        return -1  # Target not found 
```
